```python
import pandas as pd
import json
```

Math Summary File

```python
with open('/content/drive/Othercomputers/My Laptop/Learn/Youtube/PowerBi/Cricket T20/t20_json_files/t20_wc_match_results.json') as f:
  data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('/content/drive/Othercomputers/My Laptop/Learn/Youtube/PowerBi/Cricket T20/t20_csv_files/dim_match_summary.csv', index = False)
```

Batting Summary File

```python
with open('/content/drive/Othercomputers/My Laptop/Learn/Youtube/PowerBi/Cricket T20/t20_json_files/t20_wc_batting_summary.json') as f:
  data = json.load(f)

  all_records = []

  for rec in data:
    all_records.extend(rec['battingSummary'])


df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not_out')
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('/content/drive/Othercomputers/My Laptop/Learn/Youtube/PowerBi/Cricket T20/t20_csv_files/fact_bating_summary.csv', index = False)
```
